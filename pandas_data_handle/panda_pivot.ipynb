```python
import pandas as pd
import numpy as np
```

# pivot方法
piovt方法是数据重塑的一种方法，将某行变为内层索引后再执行unstack方法

```python
a = pd.DataFrame(np.arange(36).reshape([6,6]),columns=['a','b','c','d','e','f'])
a['a'] = ['x','x','x','y','y','y']
a['b'] = ['i','j','k','i','j','k']
a
```

```python
a.pivot('a','b')
```

```python
a.set_index(['a','b']).unstack()
```

# pivot_table方法
传入指定行列，可以制作数据透视表
- index=：传入要成为行索引的列
- column=：传入要成为列索引的列

```python
# print(a)
a.pivot_table(index=['a','b'],margins=True,aggfunc=len)
```

```python
a.pivot_table(columns=['a','b'])
```
