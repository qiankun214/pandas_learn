```python
import numpy as np
import pandas as pd
import json as js
import matplotlib.pyplot as plt
```

# 载入yolo2网络结构数据
## 载入数据

```python
yolo_size = [
        {"pic_size": 416, "conv_size": 3, "in_channel": 3, "out_channel": 32},
        {"pic_size": 208, "conv_size": 3, "in_channel": 32,  "out_channel": 64},
        {"pic_size": 104, "conv_size": 3, "in_channel": 64,  "out_channel": 128},
        {"pic_size": 104, "conv_size": 1, "in_channel": 128, "out_channel": 64},
        {"pic_size": 104, "conv_size": 3, "in_channel": 64,  "out_channel": 128},
        {"pic_size": 52,  "conv_size": 3, "in_channel": 128, "out_channel": 256},
        {"pic_size": 52,  "conv_size": 1, "in_channel": 256, "out_channel": 128},
        {"pic_size": 52,  "conv_size": 3, "in_channel": 128, "out_channel": 256},
        {"pic_size": 26,  "conv_size": 3, "in_channel": 256, "out_channel": 512},
        {"pic_size": 26,  "conv_size": 1, "in_channel": 512, "out_channel": 256},
        {"pic_size": 26,  "conv_size": 3, "in_channel": 256, "out_channel": 512},
        {"pic_size": 26,  "conv_size": 1, "in_channel": 512, "out_channel": 256},
        {"pic_size": 26,  "conv_size": 3, "in_channel": 256, "out_channel": 512},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 512, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 1, "in_channel": 1024, "out_channel": 512},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 512, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 1, "in_channel": 1024, "out_channel": 512},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 512, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 1024, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 1024, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 1280, "out_channel": 1024},
        {"pic_size": 13,  "conv_size": 3, "in_channel": 1024, "out_channel": 125}
    ]
```

```python
yolo_to_pd = {}
for x in range(len(yolo_size)):
    yolo_size[x]["stride"] = 1
    yolo_to_pd["conv%s" % x] = yolo_size[x]
yolo_pd_data = pd.DataFrame(yolo_to_pd).T
yolo_pd_data
yolo_pd_data = yolo_pd_data.reindex(["conv%s" % x for x in range(len(yolo_size))])
```

## 原始数据留档

```python
with open("./yolo_source.josn","w") as f:
    js.dump(yolo_to_pd,f,indent=4)
yolo_pd_data.to_csv("./yolo.csv")
```

# 计算数据
## 每层输入数据尺寸

```python
yolo_pd_data["input picture size"] = (yolo_pd_data["pic_size"] ** 2) * yolo_pd_data["in_channel"]
yolo_pd_data
```

## 每层权值尺寸

```python
yolo_pd_data["weight size"] = (yolo_pd_data.conv_size ** 2) * yolo_pd_data["in_channel"] * yolo_pd_data["out_channel"]
yolo_pd_data
```

## 运算次数

```python
yolo_pd_data["compute time"] = yolo_pd_data["weight size"] * (yolo_pd_data["pic_size"] ** 2)
yolo_pd_data
```

## 运算次数/权值数量

```python
yolo_pd_data["c-w"] = yolo_pd_data["compute time"] / yolo_pd_data["weight size"]
yolo_pd_data
```

# 权重参数/图片数据

```python
yolo_pd_data["w-p"] = yolo_pd_data['weight size'] / yolo_pd_data['input picture size']
yolo_pd_data.to_csv("./yolo.csv")
yolo_pd_data
```

```python
def basic_compute(pd_data):
    pd_data["input picture size"] = (pd_data["pic_size"] ** 2) * pd_data["in_channel"]
    pd_data["weight size"] = (pd_data.conv_size ** 2) * pd_data["in_channel"] * pd_data["out_channel"]
    pd_data["compute time"] = pd_data["weight size"] * (pd_data["pic_size"] ** 2)
    pd_data["c-w"] = pd_data["compute time"] / pd_data["weight size"]
```

# 分析
## 基本统计参数分析

```python
yolo_pd_data.describe()
```

```python
yolo_corr = yolo_pd_data.corr()
yolo_corr.reindex(["input picture size","weight size","compute time","w-p","c-w"])
```

结果：
- 对于输入数据的体量：与图片长宽呈现弱的正相关，且弱于与输出channel的负相关（由于图片在网络中流动时，尺寸变小，channel变多）
- 对于权值数据的体量：与输出channel呈很强的正相关，可以说由输入channel和输出channel共同决定
- 对于计算时间：与卷积核大小和输出channel呈正相关
