```python
import pandas as pd
import numpy as np
```

# transform
`transform()`方法传入一个函数，该函数传入一个Series，返回相同的大小的Series或标量，这个返回值会直接被放置会原有DataFrame对应的位置上(标量会广播)

```python
a = pd.DataFrame(np.random.randn(4,4))
a['group'] = ['x','x','y','y']
a
```

```python
a.groupby(['group']).transform(np.mean)
```

# apply
`apply()`是最一般的分组聚合函数，参数是一个传入Series，传出Series或标量的函数，返回值会被组合到一起

```python
def test(arr):
    return pd.concat([arr,arr])

a.groupby(['group']).apply(test)
```
